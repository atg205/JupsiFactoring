```python
mc = False
r = reader.Reader(mc=mc)
r.read_results()
df_results = r.df
```

```python
df_results
min_df_grouped = df_results[['N','min_energy',"N_dim"]].groupby('N').min().reset_index()
#min_df_grouped['N_dim'] = min_df_grouped['N'].apply(bin).apply(len)-2
min_df_grouped = min_df_grouped[['N_dim','min_energy']]
me_df = min_df_grouped.copy()
me_df['N_dim'] = min_df_grouped['N_dim'].astype('category')
me_df = me_df[me_df['min_energy']==0].groupby('N_dim').size().to_frame('count')#.reset_index(name='count')
total_df = min_df_grouped.groupby('N_dim').count().rename(columns={'min_energy':'total'})
result_df = me_df.copy()
result_df['mmt'] = me_df['count'] / total_df['total'] *100

if mc:
    filename = 'mc.pckl'
else:
    filename = 'mmt.pckl'
with open(f'/home/atg205/Documents/clone-multiplier-encoder-5/p/dfs/grd_states/{filename}','wb') as f:
    result_df.to_pickle(f)

result_df
```
