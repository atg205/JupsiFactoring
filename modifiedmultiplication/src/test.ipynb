```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [9, 10, 11, 12, 13]
}
df = pd.DataFrame(data)

# Create a figure and axes
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Create boxplots on the specified axes
df.boxplot(column='A', ax=axes[0])
axes[0].set_title('Boxplot of A')

df.boxplot(column='B', ax=axes[1])
axes[1].set_title('Boxplot of B')

df.boxplot(column='C', ax=axes[2])
axes[2].set_title('Boxplot of C')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
