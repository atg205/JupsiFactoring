```python
import sympy as sp
sp.init_printing()
```

```python

expr1 = sp.sympify('2*(1+p0*q0+q1-(4*c1+2*c0))+p0+q0-1')
expr1
```

```python
expr2 = sp.sympify('2*(q1+p0+c1-(4*c3+2*c2))+q0+p0*q1+1+c0-3')
expr2
```

```python
expr3 = sp.sympify('2*c3+1+c2-2')
expr3
```

```python
cost_expr = expr1**2+expr2**2+expr3**2
cost_expr
```

```python
p0 ,q0, q1, c0, c1, c2, c3 = sp.symbols("p0 q0 q1 c0 c1 c2 c3")
subs = [
    (p0,1),
    (q0,0),
    (q1,1),
    (c0,1),
    (c1,0),
    (c2,1),
    (c3,0)
]

cost_expr.subs(subs)
```

```python
print(sp.latex(sp.expand(cost_expr)))
```
